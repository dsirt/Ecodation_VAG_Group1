```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd 
import seaborn as sns
import matplotlib as mpl
import warnings
warnings.filterwarnings("ignore")
```

```python
exercise = pd.read_csv('data/exercise.csv', delimiter=';')
exercise.head(10)
```

```python
exercise = exercise.drop(exercise.columns[[2,4,5]], axis=1) 
plt.figure(figsize = (16,9))
ax = sns.heatmap(exercise, center = 4000)
plt.yticks(rotation = 0)
plt.show()
```

```python
fish = pd.read_csv("data/Fish.csv")
fish.head(10)
```

```python
fish = fish.drop(fish.columns[0], axis = 1)
plt.figure(figsize = (16,9))
ax = sns.heatmap(fish, center = 2000)
plt.yticks(rotation = 0)
plt.show()
```

```python
plt.figure(figsize = (16,9))
ax = sns.heatmap(fish, vmin = 0, vmax = 8000, cmap = "YlGnBu")
plt.yticks(rotation = 0)
plt.show()
```

```python
plt.figure(figsize = (16,9))
ax = sns.heatmap(fish, vmin = 0, vmax = 8000, cmap = "coolwarm")
plt.yticks(rotation = 0)
plt.show()
```

```python
plt.figure(figsize = (16,9))
ax = sns.heatmap(fish, vmin = 0, vmax = 8000, linewidths = .1)
plt.yticks(rotation = 0)
plt.show()
```

```python
plt.figure(figsize = (16,9))
ax = sns.heatmap(fish, vmin = 0, vmax = 8000, annot = True, fmt = "f")
plt.yticks(rotation = 0)
plt.show()
```

```python
plt.figure(figsize=(16,9))
ax = sns.heatmap(fish, vmin=0, vmax=8000,cmap="YlGnBu", annot=True ,fmt=".2f")
plt.yticks(rotation=0)
plt.show()
```

```python
insurance = pd.read_csv('data/insurance.csv', delimiter=',')
insurance.head(10)
```

```python
#Facet Grid
```

```python
mpl.rcParams.update(mpl.rcParamsDefault)
plt.rcParams['axes.labelsize'] = 10
plt.rcParams['xtick.labelsize'] = 10
plt.rcParams['ytick.labelsize'] = 10
```

```python
g = sns.FacetGrid(insurance, col = "sex")
g = g.map(plt.hist, "bmi", edgecolor = 'w', linewidth = 2)
```

```python
g = sns.FacetGrid(insurance, col = "sex", row = "region", height = 4, aspect = 1)
g = g.map(plt.hist, "bmi", edgecolor = 'w', linewidth = 1)
```

```python
g = sns.FacetGrid(insurance, col = "sex", row = "region", height = 3, aspect = 1)
g = g.map(plt.hist, "bmi", edgecolor = 'w', linewidth = 1)
```

```python
g = sns.FacetGrid(insurance, col = "sex", row = "region", height = 4, aspect = 1)
g = g.map(plt.hist, "bmi", color = '#F39C12', edgecolor = 'w', linewidth = 1)
```

```python
g = sns.FacetGrid(insurance, col = "region", row = "sex", height =3, aspect = 1)
g = g.map(plt.scatter, "bmi", "expenses")
```

```python
g = sns.FacetGrid(insurance, col = "region", row = "sex", height =3, aspect = 1)
g = g.map(plt.scatter, "bmi", "expenses", color = '#FF6600', edgecolor = 'w', s = 80)
```

```python
g = sns.FacetGrid(insurance, col = "region", row = "sex", hue = "smoker", height =3, aspect = 1)
g = g.map(plt.scatter, "bmi", "age", edgecolor = 'w', s = 70)
```

```python
g = sns.FacetGrid(insurance, col = "region", row = "sex", hue = "smoker", height =3, aspect = 1)
g = g.map(plt.scatter, "bmi", "expenses", edgecolor = 'w', s = 70)
g.add_legend()
```

```python
pal1 = dict(yes = "#FF7315", no = "#8CBA51")
g = sns.FacetGrid(insurance, col = "region", row = "sex", hue = "smoker", height = 3, aspect = 1, palette = pal1)
g = g.map(plt.scatter, "bmi", "expenses", edgecolor = "w", s = 70)
g.add_legend()
plt.show()
```

```python
pal1 = dict(yes = "#FF7315", no = "#8CBA51")
kws = dict(s = 80, linewidth = 1, edgecolor = "w")
kws1 = dict(height = 3, aspect = 1, palette = pal1)
g = sns.FacetGrid(insurance, col = "region", row = "sex", hue = "smoker", **kws1)
g = g.map(plt.scatter, "bmi", "expenses", **kws)
g.add_legend()
plt.show()
```

```python
pal1 = dict(yes = "#FF7315", no = "#8CBA51")
kws = dict(s = 80, linewidth = 1, edgecolor = "w")
kws1 = dict(height = 3, aspect = 1, palette = pal1, hue_kws = dict(marker = ["^", "v"]))
g = sns.FacetGrid(insurance, col = "region", row = "sex", hue = "smoker", **kws1)
g = g.map(plt.scatter, "bmi", "expenses", **kws)
g.add_legend()
plt.show()
```
