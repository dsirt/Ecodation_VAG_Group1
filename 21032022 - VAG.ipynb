```python
import numpy as np
from scipy import linalg
X = np.array([[1,3,5], [2,5,1], [2,3,8]])
print(X)
y = np.array([[10], [8], [3]])
print(y)
print("Inv matris ile yavaş çözüm")
print(linalg.inv(X).dot(y))
print("Solve ile hızlı çözüm")
print(linalg.solve(X, y))
```

```python
from scipy import integrate
x2 = lambda x : x**2
I = integrate.quad(x2, 0, 6)
print(I)
```

```python
from scipy.misc import derivative
def f(x):
    return x**3 + x**2
D = derivative(f, 1.0, dx = 1e-6)
print(D)
```

```python
#Birden fazla veri kümesini temsil etme
plt.figure(figsize=(10,5))
x = np.array([1,2,3,4,5,6])
y1 = np.power(x, 2)
y2 = np.power(x, 3)
plt.plot(x, y1, "b-", label = '$y1 = x^2$')
plt.plot(x, y2, "r-", label = '$y2 = x^3$')
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
x = np.linspace(0, 10, 2000)
plt.figure(figsize = (10,6))
plt.plot(x, np.sin(x) , label = '$Sin(X)$')
plt.plot(x, np.cos(x) , label = '$Cos(X)$')
plt.xlabel(r'$X$', fontsize = 18)
plt.ylabel(r'$Y$', fontsize = 18)
plt.title("$Sin(x) $ $ & $ $ Cos(x) $", fontsize = 14)
plt.legend(loc = 'upper right')
plt.show()
```
