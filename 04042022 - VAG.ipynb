```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd 
import seaborn as sns
import matplotlib as mpl
```

```python
corona = pd.read_csv('data/COVID-19/corona.csv', index_col = 'Date', parse_dates = True)
corona.head(10)
```

```python
plt.figure(figsize = (20, 6))
sns.lineplot(data = corona[corona['Country/Region'] == 'Italy']['Confirmed'], label = "Confirmed")
sns.lineplot(data = corona[corona['Country/Region'] == 'Italy']['Deaths'], label = "Deaths")
sns.lineplot(data = corona[corona['Country/Region'] == 'Italy']['Recovered'], label = "Recovered")
plt.title("Total Confirmed Cases / Deaths / Recovered in Italy")
plt.show()
```

```python
plt.figure(figsize = (20, 6))
sns.lineplot(data = corona[corona['Country/Region'] == 'US']['Confirmed'], label = "Confirmed")
sns.lineplot(data = corona[corona['Country/Region'] == 'US']['Deaths'], label = "Deaths")
sns.lineplot(data = corona[corona['Country/Region'] == 'US']['Recovered'], label = "Recovered")
plt.title("Total Confirmed Cases / Deaths / Recovered in USA")
plt.show()
```

```python
#Relplot
```

```python
employment = pd.read_excel("data/unemployment.xlsx")
employment.head(10)
```

```python
employment.rename(columns={'Alberta':'Unemployed'}, inplace = True)
employment.head(10)
```

```python
plt.figure(figsize = (14, 16))
sns.relplot(x = "When", y = "Unemployed", hue = "Sex", data = employment, height = 7, aspect = 1)
plt.show()
```

```python
plt.figure(figsize = (14, 16))
sns.relplot(x = "bmi", y = "expenses", hue = "smoker", data = insurance, height = 8, aspect = 1)
plt.show()
```

```python
plt.figure(figsize = (10, 10))
sns.set(rc = {'xtick.labelsize':12, 'ytick.labelsize':10, 'axes.labelsize':10, "axes.grid":False})
sns.relplot(x = "When", y = "Unemployed", hue = "Sex", col = "AgeGroup", data = employment, kind = 'line',height = 5, aspect = 0.5)
plt.show()
```

```python
sns.set(rc = {'xtick.labelsize':12, 'ytick.labelsize':12, 'axes.labelsize':12, "axes.grid":False})
sns.relplot(x = "bmi", y = "expenses", hue = "sex", col = "sex", row = 'region',data = insurance)
plt.plot()
```
