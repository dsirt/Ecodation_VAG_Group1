```python
import numpy as np
import pandas as pd
import scipy as sp
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('seaborn-darkgrid')
```

```python
plt.figure(figsize = (7,9))

barwidth = 0.25

y1 = np.array([17,18,29,40])
y2 = np.array([20,21,22,23])
y3 = np.array([5,9,11,12])

pos1 = np.arange(len(y1))
pos2 = [x + barwidth for x in pos1]
pos3 = [x + barwidth for x in pos2]

plt.bar(pos1, y1, label = 'Mevcut Biletler', width = barwidth, color = '#FF6F00')
plt.bar(pos2, y2, label = 'Satılmış Biletler', width = barwidth, color = '#FFB300')
plt.bar(pos3, y3, label = 'İptal Edilmiş Biletler', width = barwidth, color = '#F7DC6F')

plt.xlabel('Bilet Sahipleri', fontweight = 'bold')
plt.ylabel('Bilet Sayısı', fontweight = 'bold')
plt.xticks([i + barwidth for i in range(len(y1))], ['Ahmet', 'Mehmet', 'Ali', 'Osman'])

plt.legend()
plt.show()
```

```python
plt.figure(figsize = (7,9))

barwidth = 0.25

y1 = np.array([17,18,29,40])
y2 = np.array([20,21,22,23])
y3 = np.array([5,9,11,12])

pos1 = np.arange(len(y1))
pos2 = [x + barwidth for x in pos1]
pos3 = [x + barwidth for x in pos2]

plt.bar(pos1, y1, label = 'Mevcut Biletler', width = barwidth, color = '#FF6F00')
plt.bar(pos2, y2, label = 'Satılmış Biletler', width = barwidth, color = '#FFB300')
plt.bar(pos3, y3, label = 'İptal Edilmiş Biletler', width = barwidth, color = '#F7DC6F')

plt.xlabel('Bilet Sahipleri', fontweight = 'bold')
plt.ylabel('Bilet Sayısı', fontweight = 'bold')
plt.xticks([i + barwidth for i in range(len(y1))], ['Ahmet', 'Mehmet', 'Ali', 'Osman'])

for x, y in zip(pos1, y1):
    plt.text(x, y, '%d' % y, ha = 'center', va = 'bottom')
    
for x, y in zip(pos2, y2):
    plt.text(x, y, '%d' % y, ha = 'center', va = 'bottom')
    
for x, y in zip(pos3, y3):
    plt.text(x, y, '%d' % y, ha = 'center', va = 'bottom')
    
plt.title('$ Gruplanmış $ $ Bar $ $ Grafik $')

plt.legend()
plt.show()
```

```python
x1 = np.array([250, 150, 350, 252, 450, 550, 455, 358, 158, 355])
y1 = np.array([40, 50, 80, 90, 100, 50, 60, 88, 54, 45])

x2 = np.array([200, 100, 300, 220, 400, 500, 450, 380, 180, 350])
y2 = np.array([400, 500, 800, 900, 1000, 500, 600, 808, 504, 405])

#Grafik 1
plt.scatter(x1, y1)
plt.xlabel('$ Harcanan $ $ Zaman $', fontsize = 12)
plt.ylabel('$ Skor $', fontsize = 12)
plt.title('Saçılım Grafiği')
plt.show()
```

```python
#Grafik 2
plt.scatter(x2, y2, color = 'r')
plt.xlabel('$ Harcanan $ $ Zaman $', fontsize = 12)
plt.ylabel('$ Skor $', fontsize = 12)
plt.title('Saçılım Grafiği')
plt.show()
```

```python
#Grafik 3
plt.scatter(x1, y1, label = 'Class 1')
plt.scatter(x2, y2, label = 'Class 2', color = 'r')
plt.xlabel('$ Harcanan $ $ Zaman $', fontsize = 12)
plt.ylabel('$ Skor $', fontsize = 12)
plt.title('Saçılım Grafiği')
plt.legend()
plt.show()
```

```python
#Grafik 4
plt.scatter(x1, y1, label = 'Class 1', marker = 'o', color = 'm')
plt.scatter(x2, y2, label = 'Class 2', marker = 'v', color = 'r')
plt.xlabel('$ Harcanan $ $ Zaman $', fontsize = 12)
plt.ylabel('$ Skor $', fontsize = 12)
plt.title('Saçılım Grafiği')
plt.legend()
plt.show()
```

```python
#Stacked Area Plot
```

```python
x = np.arange(1, 6)
y1 = np.array([1,5,9,13,17])
y2 = np.array([2,6,10,14,16])
y3 = np.array([3,7,11,15,19])
y4 = np.array([4,8,12,16,20])

plt.figure(figsize = (8,6))
plt.stackplot(x, y1, y2, y3, y4, labels = ['Y1', 'Y2', 'Y3', 'Y4'], colors = ['#00B159', '#FFC425', '#F37735', '#BAF1A1'])
plt.legend(loc = 'upper left')
plt.show()
```
